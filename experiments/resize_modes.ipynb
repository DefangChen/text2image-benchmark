```python
from PIL import Image
import numpy as np
import cv2
import matplotlib.pyplot as plt
import sys
sys.path.append('../')
from T2IBenchmark.utils import Resizer
```

```python
img = Image.new("RGB", (128, 128), "white")
center_coordinates = (64, 64)
radius = 48
color = (0, 0, 0)
thickness = 1
np_image = cv2.circle(np.array(img), center_coordinates, radius, color, thickness)
plt.imshow(np_image.astype(np.uint8))
```

```python
# bicubic interpolation in OpenCV
resizer = Resizer("OpenCV", "bicubic", True, (16, 16))
img_resized = resizer(np_image).astype(np.uint8)
plt.imshow(img_resized.astype(np.uint8))
```

```python
# bicubic interpolation in Pillow
resizer = Resizer("PIL", "bicubic", True, (16, 16))
img_resized = resizer(np_image).astype(np.uint8)
plt.imshow(img_resized.astype(np.uint8))
```
